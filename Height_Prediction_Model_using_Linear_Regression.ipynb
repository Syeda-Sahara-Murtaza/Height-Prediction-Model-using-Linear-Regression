# **Height Prediction Model using Linear Regression**

```
#import libraries
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Convert lists to NumPy arrays before reshaping
age=np.array([12,13,14,15,16,17,18,19,20]).reshape(-1,1)
height  = np.array([110, 125, 135, 145, 155, 165, 172, 175, 178])
#initialize linear regression
model=LinearRegression()
model.fit(age,height)
#predictions
print('height prediction(linear regression)')
print('.................')
#continuos loop
while True:
  age_input=float(input('Enter age:')) # Renamed variable to avoid conflict with 'age' array
  re_age=np.array([age_input]).reshape(-1,1)
  predicted_height=model.predict(re_age) # Corrected by removing extra brackets
  print(f'predicted_height:{predicted_height[0]:.2f}cm')
  choice = input("Predict again? (y/n): ").lower()
  if choice == 'n':
    print("Program ended.")
    break
```
